```python
import pyodbc
import pandas as pd

def get_sales_data():
    conn = pyodbc.connect(
        r'DRIVER={SQL Server};SERVER=web.speed.live;DATABASE=Sanad1;Trusted_Connection=yes;'
    )

    query = """
    SELECT  
        c.CUSTOMER_B2B_ID,
        ItemId,
        itemname,
        COUNT(*) AS NumPurchases ,
        Sum(Netsalesvalue) as TotalSales 
    FROM 
        MP_Sales s
    LEFT JOIN 
        MP_Customers c
    ON 
        s.CustomerId = c.SITE_NUMBER
    Where  
        Netsalesvalue>=1 and ItemId Not Like '%XE%' and c.CUSTOMER_B2B_ID NOT Like '%AX%' 
    GROUP BY
        c.CUSTOMER_B2B_ID, ItemId,itemname
    ORDER BY
        NumPurchases DESC ,  c.CUSTOMER_B2B_ID ASC """
    
    
    df = pd.read_sql(query, conn)
    conn.close()
    return df

df= get_sales_data()
df.head()
```

#  Item-Based Recommendation

```python
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
from scipy.sparse import csr_matrix

# Pivot for user-item matrix
user_item = df.pivot_table(index='CUSTOMER_B2B_ID', columns='ItemId', values='NumPurchases', fill_value=0)
user_sparse = csr_matrix(user_item)

# Transpose for item-based
item_item = user_item.T
item_sparse = csr_matrix(item_item)

# Compute similarity
item_sim = cosine_similarity(item_sparse)
item_sim_df = pd.DataFrame(item_sim, index=item_item.index, columns=item_item.index)

# Recommend similar items
def recommend_items(item_id, top_n=5):
    if item_id not in item_sim_df:
        return []
    return item_sim_df[item_id].sort_values(ascending=False)[1:top_n+1].index.tolist()
```

```python
def getItemName(item_id):
    if item_id in Preprocessed_data['ItemId'].values:
        return Preprocessed_data[Preprocessed_data['ItemId'] == item_id]['itemname'].values[0]
    else:
        return f"{item_id} not found in data"
```

```python
def get_item_info():
    conn = pyodbc.connect(
        r'DRIVER={SQL Server};SERVER=web.speed.live;DATABASE=Sanad1;Trusted_Connection=yes;'
    )

    query = """
    SELECT  
        ITEM_CODE AS ItemId,
        DESCRIPTION AS itemname
    FROM 
        MP_Items

    Where  
          ITEM_CODE  Not Like '%XE%' """
    
    
    df = pd.read_sql(query, conn)
    conn.close()
    return df

item_info= get_item_info()
item_info
```

```python
# train_model.py
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
from scipy.sparse import csr_matrix
import joblib


# Create user-item matrix
user_item_matrix = df.pivot_table(index='CUSTOMER_B2B_ID', columns='ItemId', values='NumPurchases', fill_value=0)
item_user_matrix = user_item_matrix.T

# Compute similarity
sparse_matrix = csr_matrix(item_user_matrix.values)
similarity = cosine_similarity(sparse_matrix)

# Save similarity matrix and item metadata
similarity_df = pd.DataFrame(similarity, index=item_user_matrix.index, columns=item_user_matrix.index)
joblib.dump(similarity_df, "item_similarity.pkl")
joblib.dump(item_info.set_index("ItemId").to_dict()["itemname"], "item_name_map.pkl")
```

```python
import pandas as pd
from sklearn.neighbors import NearestNeighbors
from scipy.sparse import csr_matrix

# Step 1: Load and prepare data
# Assume 'matrix' has columns: CUSTOMER_B2B_ID, ItemId, NumPurchases
matrix = df[['CUSTOMER_B2B_ID', 'ItemId', 'NumPurchases']]

# Step 2: Create user-item matrix
user_item_matrix = matrix.pivot_table(index='CUSTOMER_B2B_ID', columns='ItemId', values='NumPurchases', fill_value=0)

# Step 3: Convert to sparse matrix
sparse_matrix = csr_matrix(user_item_matrix.values)

# Step 4: Fit NearestNeighbors model
model = NearestNeighbors(metric='cosine', algorithm='brute', n_neighbors=6, n_jobs=-1)
model.fit(sparse_matrix)

# Step 5: Pick a target user to recommend for
target_user_index = 0  # example: first user in the matrix
distances, indices = model.kneighbors(sparse_matrix[target_user_index], n_neighbors=6)

# Note: indices[0][0] is the user themselves
similar_user_indices = indices[0][1:]  # exclude the user themselves

# Step 6: Get items purchased by similar users but not the target
target_user_id = user_item_matrix.index[target_user_index]
similar_user_ids = user_item_matrix.index[similar_user_indices]

# Items the target user already interacted with
target_user_items = set(user_item_matrix.loc[target_user_id][user_item_matrix.loc[target_user_id] > 0].index)

# Items from similar users
similar_users_matrix = user_item_matrix.loc[similar_user_ids]
recommendation_scores = similar_users_matrix.sum(axis=0)

# Drop items the user has already interacted with
recommendation_scores = recommendation_scores.drop(labels=target_user_items)

# Top recommended items
top_items = recommendation_scores.sort_values(ascending=False).head(10)
print("Top recommended items for user", target_user_id)
print(top_items)
```

```python
def recommend_items_for_user(target_user_b2b_id, user_item_matrix, model, top_k=5, top_n_items=10):
    import numpy as np

    if target_user_b2b_id not in user_item_matrix.index:
        return f"User {target_user_b2b_id} not found in data."
    
    # Get the index of the user
    target_index = user_item_matrix.index.get_loc(target_user_b2b_id)

    # Find K nearest neighbors (including the user)
    distances, indices = model.kneighbors([user_item_matrix.iloc[target_index].values], n_neighbors=top_k + 1)

    # Get similar user IDs (excluding the user)
    similar_user_indices = indices[0][1:]
    similar_user_ids = user_item_matrix.index[similar_user_indices]

    # Items the user already interacted with
    target_user_items = set(user_item_matrix.loc[target_user_b2b_id][user_item_matrix.loc[target_user_b2b_id] > 0].index)

    # Aggregate interactions from similar users
    similar_users_matrix = user_item_matrix.loc[similar_user_ids]
    recommendation_scores = similar_users_matrix.sum(axis=0)

    # Filter out already seen items
    recommendation_scores = recommendation_scores.drop(labels=target_user_items, errors='ignore')

    # Top N recommended items
    top_items = recommendation_scores.sort_values(ascending=False).head(top_n_items)

    return top_items
```

```python
# Step 1: Create your user_item_matrix (same as before)
user_item_matrix = matrix.pivot_table(index='CUSTOMER_B2B_ID', columns='ItemId', values='NumPurchases', fill_value=0)

# Step 2: Train the model once
sparse_matrix = csr_matrix(user_item_matrix.values)
model = NearestNeighbors(metric='cosine', algorithm='brute', n_neighbors=6, n_jobs=-1)
model.fit(sparse_matrix)

# Step 3: Recommend for a specific user B2B ID
target_b2b_id = '132'  # replace with your actual user ID
recommendations = recommend_items_for_user(target_b2b_id, user_item_matrix, model, top_k=5, top_n_items=5)

print(f"Top recommended items for user {target_b2b_id}:")
print(recommendations)
```
